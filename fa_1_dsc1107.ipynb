```r
# Load required packages (ggplot2 is part of tidyverse)
library(tidyverse)

# View the dataset
glimpse(diamonds)

# Boxplot of carat by cut
ggplot(diamonds, aes(x = cut, y = carat)) +
  geom_boxplot() +
  labs(title = "Diamond carat weight by cut quality",
       x = "Cut", y = "Carat")

# Optional: violin plot for more detail
ggplot(diamonds, aes(x = cut, y = carat)) +
  geom_violin() +
  labs(title = "Distribution of carat by cut", x = "Cut", y = "Carat")
```

The plot shows that diamonds with better cuts (Ideal, Premium) tend to be smaller on average, while those with poorer cuts (Fair, Good) are heavier. This explains the paradox seen in the price‑by‑cut boxplot: although better cuts are generally more valuable, the diamonds in those categories are smaller, and because price is strongly driven by carat, the median price for Ideal ends up being lower than for Fair.

```r
# Scatter plot with zoom
ggplot(diamonds, aes(x = carat, y = x)) +
  geom_point(alpha = 0.5, size = 0.8) +          # alpha reduces overplotting
  coord_cartesian(xlim = c(0, 3), ylim = c(3, 10)) +
  labs(title = "Diamond length vs. carat (zoomed in)",
       x = "Carat", y = "Length (x, mm)")

# Optional: add a linear trend line for clarity
ggplot(diamonds, aes(x = carat, y = x)) +
  geom_point(alpha = 0.3, size = 0.8) +
  geom_smooth(method = "lm", se = FALSE, color = "red") +
  coord_cartesian(xlim = c(0, 3), ylim = c(3, 10)) +
  labs(title = "Linear relationship between carat and length",
       x = "Carat", y = "Length (x, mm)")
```

After excluding outliers, the plot shows a strong, roughly linear positive relationship: as carat increases, the physical length (x) increases. This is expected because larger diamonds (higher carat weight) have greater dimensions. The relationship is not perfectly linear because diamonds are cut to different proportions, but over the common range, a linear approximation works well.

why it fails: The column names 1999 and 2000 are numbers, not syntactic names. Writing c(1999, 2000) tells R to look for columns by numeric position (1, 2, 3…). Since there are only 3 columns, positions 1999 and 2000 don't exist.
