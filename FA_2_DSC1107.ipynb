```r
library(tidyverse)

# 2.1 Preview the (built‑in) datasets
head(who)
glimpse(who)
head(population)
glimpse(population)
# ----------------------------------------------------------------------
# Instead of importing from files, load the built-in datasets
# (they come with the tidyverse, no upload needed)
# ----------------------------------------------------------------------
data("who",          package = "tidyr")   # same messy wide format as who.tsv
data("population",   package = "tidyr")   # already tidy: country, year, population

# 2.3 Dimensions
dim(who)
dim(population)

# 2.4 The built-in "population" is already tidy and numeric,
#     but we'll just store a clean copy as "population2" to match the instructions.
population2 <- population

# ----------------------------------------------------------------------
# 3. Tidy Data
# ----------------------------------------------------------------------

# --- 3.1 who dataset ---
# 3.1.2 Pivot to long
who2 <- who %>%
  pivot_longer(
    cols = starts_with("new_"),
    names_to = "key",
    values_to = "cases"
  )

# 3.1.3 Separate into new, type, sex_age; remove 'new'
who3 <- who2 %>%
  separate(key, into = c("new", "type", "sex_age"), sep = "_") %>%
  select(-new)

# 3.1.4 Split sex_age into sex and age_group
who_tidy <- who3 %>%
  separate(sex_age, into = c("sex", "age_group"), sep = 1)

# --- 3.2 population dataset ---
# The built-in population is already in long form (country, year, population).
# We just need to rename/ensure types to create population_tidy as required.
population_tidy <- population2 %>%
  rename(population = population) %>%      # already named 'population' in the dataset
  mutate(
    year = as.integer(year),
    population = as.numeric(population)
  )

# --- 3.3 Join datasets ---
tuberculosis <- left_join(who_tidy, population_tidy, by = c("country", "year"))

# --- 3.4 Clean up ---
# 3.4.1 Remove unnecessary columns
tuberculosis <- select(tuberculosis, -iso2, -iso3)

# 3.4.2 Drop rows with missing cases or population
tuberculosis <- filter(tuberculosis, !is.na(cases), !is.na(population))

# (3.4.3 is already satisfied – the cleaned data stays in 'tuberculosis')

# ----------------------------------------------------------------------
# 4. Data Manipulation
# ----------------------------------------------------------------------
# 4.1 Total TB cases among men/women in USA, 21st century
us_21c <- tuberculosis %>%
  filter(country == "United States", year >= 2000) %>%
  group_by(sex) %>%
  summarise(total_cases = sum(cases, na.rm = TRUE))

us_21c   # inspect to see which sex has more cases

# 4.2 Cases per 100k
tuberculosis <- tuberculosis %>%
  mutate(cases_per_100k = (cases / population) * 100000)

# 4.3 Highest and lowest overall rate (by country and year)
country_year_rate <- tuberculosis %>%
  group_by(country, year) %>%
  summarise(
    total_cases = sum(cases, na.rm = TRUE),
    total_pop   = first(population),   # population constant for a country-year
    .groups = "drop"
  ) %>%
  mutate(rate_per_100k = (total_cases / total_pop) * 100000)

country_year_rate %>% slice_max(rate_per_100k, n = 1)
country_year_rate %>% slice_min(rate_per_100k, n = 1)

# ----------------------------------------------------------------------
# 5. Data Visualization
# ----------------------------------------------------------------------
# 5.1 Time series for China, India, USA (log scale)
three_countries <- country_year_rate %>%
  filter(country %in% c("China", "India", "United States"))

ggplot(three_countries, aes(x = year, y = rate_per_100k, color = country)) +
  geom_line(size = 1) +
  scale_y_log10() +
  labs(title = "TB cases per 100,000 population",
       y = "Rate per 100k (log scale)", x = "Year") +
  theme_minimal()

# 5.2 Distribution by age group (log scale)
age_group_rate <- tuberculosis %>%
  group_by(country, age_group) %>%
  summarise(
    total_cases = sum(cases, na.rm = TRUE),
    total_pop   = sum(population, na.rm = TRUE),
    .groups = "drop"
  ) %>%
  mutate(rate_per_100k = (total_cases / total_pop) * 100000)

ggplot(age_group_rate, aes(x = age_group, y = rate_per_100k)) +
  geom_boxplot() +
  scale_y_log10() +
  labs(title = "TB rate per 100k by age group",
       y = "Rate per 100k (log scale)", x = "Age group") +
  theme_minimal()

# 5.3 TB rate in 2000 vs. population
data_2000 <- country_year_rate %>% filter(year == 2000)

ggplot(data_2000, aes(x = total_pop, y = rate_per_100k)) +
  geom_point(alpha = 0.5) +
  geom_smooth(method = "loess", se = FALSE) +
  scale_x_log10() +
  labs(title = "TB rate in 2000 vs. country population",
       x = "Population (log scale)", y = "Cases per 100k") +
  theme_minimal()
```

**Conclusion**

No clear 'linear' pattern of behaviour is evident from the scatterplot. It does, however, seem that on average smaller countries have either extremely low (or extreme high - spread out of values) numbers of cases of TB; but that more populated countries usually have moderate amounts of TB reported. The overall conclusion is that the size of a country's population in the year 2000 does not appear to significantly predict its burden of TB; there are numerous other factors influencing the incidence of TB, including, but not limited to public health infrastructure, high levels of poverty, and prevalence of HIV.
